```python
# WAP to findout min and max value from a list

l=[10,20,4,7,2,9]

min=l[0]

for i in l:
    if min>i:
        min=i
print(f'min is {min}')
```

```python
# WAP to find out sum of all elements of list and also find out mean

l=[10,20,30,40,50]

sum=0

for i in l:
    sum=sum+i

print('sum of all no is',sum)

mean=sum/len(l)

print('mean is',mean)
```
