```python
#list

l=[23,33,'Abakash',True,9.9]
print(type(l))
print(l)
```

```python
#accessing element through index

l=[23,33,'Abakash',True,9.9]

print(l[0])
print(l[2])
print(l[-2])
print(l[5])
```

```python
# creation of list using various way
```

```python
# creation of empty list
l=[]
print(l)
```

```python
# create of list using eval()

l=eval(input('enter list elements :'))

print(l)
```

```python
# creation of list using list() function

l=list((10,20,21,22,23))
print(l)
```

```python
#creation of list using range

l=list(range(10,21))
print(l)
```

```python
# creation of list directly

l=[23,33,'Abakash',True,9.9]
print(l)
```

```python
# heterogenous in nature

l=[10,29,('ab')]
```

```python
# creation of list using split function

msg='hey welcome'
l=msg.split()

print(l)
```

```python
# accessing element using slice operator

l=[23,33,43,5,67]

print(l[: :])
print(l[: :2])
print(l[2: :2])
```

```python
# traversing of list using for loop
l=[23,33,43,54,65]

for i in l:
    print(i)
```

```python
# traversing using while loop

l=[23,33,34,54,67,87]

i=0
while i<=4:
    print(l[i])
    i=i+1
```

```python
l=[23,33,34,54,67,87]
i=0

while i<len(l):
    print(l[i])
    i=i+1
```

```python
# WAP to display list element of a list along with positive and negative index using for loop

l=[10,20,30,40]

x=len(l)

for i in range(x):
    print(f'{l[i]} is present at index {i}/{i-x}')
```
