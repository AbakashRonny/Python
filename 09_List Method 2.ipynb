```python
#shallow copy(l.copy)
```

```python
l=[23,33,43,78,89]
print(l)
s=l.copy()
print(s)
```

```python
l=[23,33,43,78,89]
print(l)
s=l.copy()
print(s)

print(id(l))    
print(id(s))

print(id(l[0]))   
print(id(s[0]))
```

```python
# deep copy(=)
```

```python
l=[23,33,43,78,89]
print(l)
s=l           
print(s)

print(id(l))    
print(id(s))

print(id(l[0]))   
print(id(s[0]))
```

```python
# updation in list in case of (=) it will reflect to another list also
```

```python
l=[23,33,43,78,89]
print(l)
s=l           
print(s)

l[2]=76
print(l)

print(id(l))
print(s)
print(id(s))

print(id(l[0]))   
print(id(s[0]))
```

```python
 # updation in list in case of (l.copy) it will not reflect to another list
```

```python
l=[23,33,43,78,89]
s=l.copy()          

l[2]=76
print(l)

print(id(l))
print(s)
print(id(s))

print(id(l[0]))   
print(id(s[0]))
```
