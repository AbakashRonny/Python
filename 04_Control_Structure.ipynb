```python
# choice 
```

```python
num1=int(input('enter no 1:'))
num2=int(input('enter no 2:'))
choice=input('enter symbol[+ - * /]')

if choice=='+':
    print(num1+num2)
elif choice=='-':
    print(num1-num2)
elif choice=='*':
    print(num1*num2)
elif choice=='/':
    print(num1/num2)
else:
    print('Not a Valid Symbol')
```
